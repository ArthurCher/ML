```python
X = data.drop(columns=['Cover_Type'])
y = data['Cover_Type']
```

Самые большие значения корреляции:

В остальных случаях коэффициент корреляции меньше 0,5 - я думаю это слабая корреляция. И учитывать ее не будем.\
Исходя из списка выше по очередно удалим признаки: Hillshade_3pm, Horizontal_Distance_To_Hydrology, Soil_Type_28, построим на каждом шаге модель и померим качество. А перед этим простроим модель на всех данных и посчитаем качество на всех данных\
Думаю, что в данном случае не очень хорошее решение делать тепловую карту матрицу корреляции - она очень мелкая, не удобно ее смотреть

```python
from sklearn.linear_model import LogisticRegression # она почему-то считала быстрее
from sklearn.model_selection import train_test_split
```

```python
# Получим первоначальные данные обучения при полном датафрейме
list_score = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) # Делим выборки

model_LA = LogisticRegression()

model_LA.fit(X_train, y_train)

first_score = model_LA.score(X_test, y_test)

print("Первоначальное качество модели ->", first_score)

list_score.append(first_score)
```

Уберем поочередно самые коррелирующие параметры и на каждом шаге будем обучать модель и замерять ее качество

```python
list_Feature = ['Hillshade_3pm', 'Horizontal_Distance_To_Hydrology', 'Soil_Type_28']
X_clean = X

for feature in list_Feature:
    X_clean = X_clean.drop(columns=[feature])
    
    X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.25, random_state=42) # Делим выборки

    model_LA = LogisticRegression()

    model_LA.fit(X_train, y_train)

    score = model_LA.score(X_test, y_test)

    print("Качество модели без параметра ", feature, "->", score, "\nОсталось параметров ->", len(X_clean.columns))

    list_score.append(score)
    
```

```python
plt.plot(list_score)
```

Как видно качество падает при удалении параметров на основе корреляции. И не удивительно - корреляция по-моему все равно слабая, чтобы на ее основе делать выводы и убирать параметры.

Отберем признаки на основе feature_impotance

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()

model.fit(X, y)

model.feature_importances_

list_importances = []

for i in range(len(list(X.columns))):
    list_importances.append(model.feature_importances_[i])
```

Переберем параметры начиная от самого важного и заканчивая самым неважным. На каждом шаге обучаем модель и измеряем качество

```python
X_clean = pd.DataFrame()

list_score = [first_score]

for feature in df_feature['feature']:
    X_clean[feature] = X[feature]
    
    X_train, X_test, y_train, y_test = train_test_split(X_clean, y, test_size=0.25, random_state=42) # Делим выборки

    model_LA = LogisticRegression()

    model_LA.fit(X_train, y_train)

    score = model_LA.score(X_test, y_test)

    print("Качество модели с параметром ", feature, "->", score, "\nВсего параметров ->", len(X_clean.columns))

    list_score.append(score)
    
```

```python
plt.figure(figsize=(12,9))
plt.plot(list_score)
```

Самое лучшее качество получается с двумя параметрами Horizontal_Distance_To_Roadways и Elevation - 0.6293088610906488. Это выше стартовых значений

Теперь перейдем к жадным алгоритмам. \
Начнем с жадного отбора признаков по максимальному нарастанию качества модели\

Т.к. мне придется перебирать все параметры попарно, на каждом шаге обучать модель и сравнивать качество выбирая вариант с самым лучшим - на датафрейме в 500 000 объектов это будет считаться очень долго.\
Поэтому для ускорения я решил взять выборку размеров в 5810 объектов.\
Чтобы сохранить пропорции классов, использую StratifiedShuffleSplit\
Для отбора параметров для обучения по всем данным - я думаю это преемлемое решение

```python
from sklearn.model_selection import StratifiedShuffleSplit

split_list = []

sss = StratifiedShuffleSplit(n_splits=1, train_size=0.01, random_state=42)
for train_id, test_id in sss.split(X, y):
    X_split = X.iloc[train_id]
    y_split = y.iloc[train_id]
```

```python
"""
Здесь реализован жадный отбор параметров: я на первом шаге перебираю параметры по очереди
Лучший параметр добавляю в список и уже на следующем шаге перебираю комбинации этого параметра с остальными. 
Параметр, который в комбинации с параметром из первого шага, показал лучший результат - добавляю в список
И т.д. пока модель улучшается.
"""

best_feature_list = []
best_score_list = []
step_list = []

X_clean = pd.DataFrame()
full_features = list(X.columns)

best_score = 0

while full_features != 0:
    best_feature = ''
    for feature_level1 in full_features:
        step_list = best_feature_list
        step_list.append(feature_level1)
        print("Проверяется свойство ->", step_list)
        
        X_clean = X_split[step_list]
        X_train, X_test, y_train, y_test = train_test_split(X_clean, y_split, test_size=0.25, random_state=42) # Делим выборки

        model_LA = LogisticRegression()

        model_LA.fit(X_train, y_train)

        score = model_LA.score(X_test, y_test)
        
        print("Качество модели ->", score)

        if best_score < score:
            best_score = score
            best_feature = feature_level1
        
        print("Лучшее показатель качества на данный момент ->", best_score)
        step_list.remove(feature_level1)
    if best_feature != '':
        best_feature_list.append(best_feature)
        best_score_list.append(best_score)
        print("---------------------------------------------------------------------")
        print("Список параметров ->", best_feature_list)
        print("Качество модели ->", best_score)
        print("---------------------------------------------------------------------")
        full_features.remove(best_feature)
    
    else:
        print("---------------------------------------------------------------------")
        print("Список параметров ->", best_feature_list)
        print("Качество модели ->", best_score)
        print("---------------------------------------------------------------------")
        break
        
```

```python
plt.figure(figsize=(12,9))
plt.plot(best_score_list)
```

Лучшее качество показало обучение на следующих параметрах: ['Wilderness_Area_3', 'Elevation', 'Soil_Type_19', 'Soil_Type_12', 'Soil_Type_20', 'Soil_Type_23', 'Soil_Type_3', 'Wilderness_Area_1'] \
Качество модели -> 0.6779077770130764\
Заметно выше стартовых значений

```python
"""
жадное отбрасывание признаков по минимальному падению качества модели
У меня есть стартовое значение качества модели. 
Я итерационно на каждом шаге исключаю по очереди один из параметров и считаю дельту со стартовым значением
Сравниваю с предыдущей дельтой и запоминаю параметр и качество модели, при которой эта дельта меньше всего
Потом этот параметр убираю и повторяю шаги
И так пока не переберу все параметры.
"""

worst_feature_list = []
best_score_list = []
step_list = []

X_clean = pd.DataFrame()
full_features = list(X.columns)
# best_score = 0
best_score = first_score

while full_features != 0:
    worst_feature = ''
    min_delta = best_score
    for feature_level1 in full_features:
        step_list = full_features.copy()
        step_list.remove(feature_level1)
        
        if len(step_list) != 0:
            print("Проверяется свойство ->", feature_level1)
            X_clean = X_split[step_list]
            X_train, X_test, y_train, y_test = train_test_split(X_clean, y_split, test_size=0.25, random_state=42) # Делим выборки

            model_LA = LogisticRegression()

            model_LA.fit(X_train, y_train)

            score = model_LA.score(X_test, y_test)

            print("Качество модели ->", score)

            if min_delta > best_score - score:
                min_delta = best_score - score
                best_score = score
                worst_feature = feature_level1

            print("Лучшее показатель качества на данный момент ->", best_score)
            print("Количество параметров", len(full_features))

    if len(step_list) > 0:
        worst_feature_list.append(worst_feature)
        best_score_list.append(best_score)
        print("---------------------------------------------------------------------")
        print("Список параметров ->", worst_feature_list)
        print("Качество модели ->", best_score)
        print("---------------------------------------------------------------------")
        full_features.remove(worst_feature)
    
    else:
        break
        
```

```python
plt.figure(figsize=(12,9))
plt.plot(best_score_list)
```

Самый лучший результат оказался когда поочередно убрали следующие параметры: ['Aspect', 'Slope', 'Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways', 'Hillshade_9am', 'Hillshade_3pm', 'Hillshade_Noon', 'Elevation', 'Horizontal_Distance_To_Fire_Points', 'Soil_Type_38', 'Soil_Type_30', 'Soil_Type_21']. Качество стало 0.6455609084652443, затем оно стало постепенно снижаться.
